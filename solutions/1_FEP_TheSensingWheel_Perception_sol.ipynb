## Part a: Variational Free Energy (VFE) - with the Laplace approximation

$ F = KL[q(x)|| p(x|s)] -ln p(s) \rightarrow_\text{laplace} F = -\ln p(s,\mu) - \frac{1}{2}(\ln |\Sigma| + \ln 2\pi)$ 


Perception is defined as the minimization of $F$:

 $\mu = \arg\min_s F(\mu, s)$
 
We compute it using the following differential equation (gradiend optimization):

 $ \dot{\mu} = -\frac{\partial F}{\partial \mu} = - \frac{\partial \ln p(s|\mu) p(\mu)}{\partial \mu}$ (Eq. 1)

 Assuming that the sensors measurements $s$ are defined by a generative model with Gaussian noise $z$:  $s = g(\mu) + z$

 We have:

 $ \dot{\mu} = \frac{\partial g(\mu)}{\partial \mu}^T \Sigma_s^{-1} (s-g(\mu)) $ (Eq. 2)

### Write down the derivation from (Eq. 1) to (Eq. 2)

1. The likelihood $p(s|\mu)$ that defines the generative model follows a Normal distribution with mean $\mu$ and $\sigma$ standard deviation:
$p(s|mu) \sim \mathcal{N}(s; \mu, \sigma^2)$ 

2. The prior $p(\mu)$ is assumed to be uniform

*Tip: use the function of the Normal density function (pdf). [wiki](https://en.wikipedia.org/wiki/Normal_distribution)*

<font color='red'>Your derivation here:</font>

We compute the partial derivative of F with respect to $\mu$. This is 

$\dot{\mu} = -\frac{\partial \ln p(s|\mu) p(\mu)}{\partial \mu}$     (Eq. 3)

We know that 

* $p(s|\mu) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{1}{2}\frac{(s-g(\mu))^2}{\sigma^2}}$

* $p(\mu) \sim \mathcal{U}(0,1)$ so we can take it out from the equation

Substituting in the equation 3 (*notation: $\partial_\mu f = \frac{\partial f}{\partial \mu}$*)


$- \partial_\mu \ln \left[ \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{1}{2}\frac{(s-g(\mu))^2}{\sigma^2}}\right] = - \left[   \partial_\mu \ln \frac{1}{\sqrt{2\pi\sigma^2}} + \partial_\mu \frac{1}{2}\frac{(s-g(\mu))^2}{\sigma^2} \right]$


The first term does not depend on $\mu$. Thus, we can take it out. Computing the partial derivative of the second summation term we have:

$ \dot{\mu} =  \frac{g(\mu)^T}{\partial \mu} \frac{(s-g(\mu))}{\sigma^2} = \frac{g(\mu)^T}{\partial \mu} \Sigma^{-1} (s-g(\mu)) $

Note that the $\frac{1}{2}$ is eliminated due to the exponential term when doing the derivative and that $\sigma^2$ is a constant. The minus becomes plus due to the minus in  $-1 * g(\mu)$. $\Sigma^{-1}$ is the inverse of the variance.

It is important to highlight that $(s-g(\mu))$ is the prediction error.


## Part b: Generative model definition

```python
import numpy as np
import matplotlib.pyplot as plt

# Generative model definition
def g(x):     # forward model
  return x**2

def dg(x):    # inverse model
  return 2*x
```

```python
# Plotting actual and inferred position of the agent

def plot_results(real_x, mu_0):

    x = np.linspace(-2, 2, 1000)
    y = np.linspace(2, 2, 1000)
    plt.plot(x, y, 'k')
    plt.plot(real_x, 2, 'r8', markersize=15)
    plt.plot(mu_0, 2, 'b8', markersize=15)
    plt.plot(0, 2, 'k|', markersize=8)
    
    plt.xlim([-1,1])  # adjust plot when choosing values outside of range -1 and 1
    plt.legend(['Space', 'actual position', 'estimated position', 'x=0'])
    plt.xticks([])
    plt.yticks([])
    plt.title('Estimated vs. actual position')

    plt.show()
```

## Initialization

```python
dt = 0.001    # integration step
ftime = 1   # maximum time considered

time = np.arange(0, ftime, dt) # Time array
N = time.shape[0] # Steps

real_x = 0.1    # real position of the ball
real_x_dot = 0  # change on the position 

mu_0 = 0.5      # internal belief 
mu_dot = 0.0     # change on the internal belief

sigma_s = 0.01    # variance of the sensory process
```

### How we compute the real sensor measurement?

```python
s = g(real_x) # we assume that the real and the brain generative function is the same.
print(s)
```

### How we compute the predicted sensor value depending on the brain variable $\mu$?

```python
s_hat = g(mu_0) # write here the code
print(s_hat)
```

### How we compute the predicion error?

```python
e_s = s-s_hat
print(e_s)
```

## Part c: implement the FEP model

```python

def fep(real_x, mu_0):
    data = [] # for debugging

    mu_dot = 0.0     # change on the internal belief
    real_x_dot = 0.0  # change on the position 

    for i in range(0,N):
        # observations (sensors)
        # get the sensor measurement we will use the g function (here we assume that the real and the brain generative model is the same)
        s = g(real_x)

        # Compute the forward model - predictor From internal state (mu_0 latent space) to sensory space
        s_hat = g(mu_0)

        # Compute the prediction error 
        e_s = s - s_hat 

        # Update brain variable according to the function in the explanation
        mu_dot =  dg(mu_0)* 1/sigma_s * e_s  
        mu_0 = mu_0 + dt * mu_dot # integrate the differential equation using 1st order Euler

        # Real agent simulation
        real_x_dot = 0 # it is not moving - no change in the position
        real_x = real_x + dt*real_x_dot

        if i%(ftime/dt/20) == 0:
            print('Iteration:', i)
            plot_results(real_x, mu_0)


        #data store for plotting
        data.append(
                [s, # 0
                real_x, # 1
                mu_0, # 2
                e_s  # 3
                ]
            )
    return data

# Initialization 
real_x = 0.1    # real position of the ball
mu_0 = 0.5      # internal belief 

data = fep(real_x, mu_0)
```

## Part d: inspect the results

```python

def plot_variables(data):
    # Plot variables
    data = np.array(data).T # transform for ploting
    print(time.shape, data.shape)

    # visualising 
    plt.figure(figsize=(10, 10))
    plt.subplots_adjust(hspace=0.3)

    ax1 = plt.subplot(311)
    ax1.plot( time, data[0, :] )
    ax1.legend(['s'])
    ax1.title.set_text('Sensory measurment')

    ax2 = plt.subplot(312)
    ax2.plot(time, data[1, :], 'b' )  
    ax2.plot(time, data[2, :], 'r' )  
    ax2.legend(['real_x', 'mu_0'])
    ax2.title.set_text('Actual vs. assumed position')

    ax3 = plt.subplot(313)
    ax3.plot(time, data[3, :] , 'k' )  
    ax3.legend(['e_s'])
    ax3.set_xlabel('time')
    ax3.title.set_text('Prediction error')

    plt.show()
    
plot_variables(data)
```

## Now add different levels of Gaussian noise to the sensor measurements and run again the FEP model

Write a function fep2 with an extra parameter that is the measurement standard deviation $\sigma$, and plot the variables

$s = g(x) + N(0,\sigma^2)$ - Increase $\sigma$ to increase the input noise

```python
def fep2(real_x, mu_0, measurement_std, sigma_s):
    data = [] # for debugging

    mu_dot = 0.0     # change on the internal belief
    real_x_dot = 0.0  # change on the position 

    for i in range(0,N):
        # observations (sensors)
        # get the sensor measurement we will use the g function (here we assume that the real and the brain generative model is the same)
        s = g(real_x) + np.random.normal(0,measurement_std)

        # Compute the forward model - predictor From internal state (mu_0 latent space) to sensory space
        s_hat = g(mu_0)

        # Compute the prediction error 
        e_s = s - s_hat 

        # Update brain variable according to the function in the explanation
        mu_dot =  dg(mu_0)* 1/sigma_s * e_s  
        mu_0 = mu_0 + dt * mu_dot # integrate the differential equation using 1st order Euler

        # Real agent simulation
        real_x_dot = 0 # it is not moving - no change in the position
        real_x = real_x + dt*real_x_dot

        if i%(ftime/dt/20) == 0:
            print('Iteration:', i)
            plot_results(real_x, mu_0)


        #data store for plotting
        data.append(
                [s, # 0
                real_x, # 1
                mu_0, # 2
                e_s  # 3
                ]
            )
    return data


real_x = 0.1    # real position of the ball
mu_0 = 0.5      # internal belief 
measurement_std = 0.1
sigma_s = 0.01    # variance of the sensory process this is for the next exercise

data = fep2(real_x, mu_0, measurement_std, sigma_s)
plot_variables(data)
```

### How does the system deals with the noise? and what means $\sigma_s$ for the agent in normal words?

<font color='red'>Describe your findings here</font>

Only when the input noise (sensor measurement) and the weighted variance are in the same order of magnitud, the system is able to porperly infer the postition. In fact this model works as an smoothing function, similar to a Kalman filter. In the case that the $sigma_s$ is too small in compare with the measurement noise the estimation will oscilate. Otherwise, it will not be able to track the changes. $sigma_s$ describes the relevance of the error to infer the internal state. When the input is noisy it is better to not use that signal.

### Now describe how you would implement the action in the FEP mathematical framework

 <font color='red'>Describe it here, you can use equations</font>
 
 The action follows the same principle as the perception. We minimize the VFE:
 
 $a = \arg\min_a F(\mu, s)$
 
 In this case $s$ is a function of $a$: $s(a)$
 
We compute it using the following differential equation (gradiend optimization):

 $\dot{a} = -\frac{\partial F}{\partial a}$
 
We can apply the chain rule and then:

 $\dot{a} = -\frac{\partial s}{\partial a} \frac{\partial F}{\partial s}$
 
Where $\frac{\partial s}{\partial a}$ is the change on the sensory input given the action
